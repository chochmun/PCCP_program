### 재귀적 이진 탐색

image.png

```python
def binary_search(L, x):
    # L.sort() # 정렬되지 않았다면
    lower = 0
    upper = len(L) - 1
    idx = -1
    while lower <= upper:
        middle = (lower + upper) // 2
        if L[middle] == x:
            idx = middle
            return idx
        elif L[middle] < x:
            lower = middle + 1
        else:
            upper = middle - 1
    return idx  # 요소를 찾지 못한 경우 -1을 반환

L = [2, 3, 6, 7, 8, 9, 10]
print(binary_search(L, 6))  # 6이 있는 인덱스를 출력
print(binary_search(L, 1))  # 1이 리스트에 없으므로 -1을 출력
```

# 연결 리스트(Linked Lists)

## 연결 리스트(Linked List) 란?

image.png

- 배열과 함께 가장 기본이 되는 대표적인 선형 자료구조 중 하나로 추상 자료형 구현의 기반이 됨

- 연결 리스트는 데이터의 순서가 메모리에 물리적인 순서대로 저장되지는 않고 각 요소를 포인터로 연결하여 관리하는 선형 자료구조

- 각 요소는 노드인 데이터 영역과 포인터 영역으로 구성

- 배열의 공간낭비를 피할 수 있는 자료구조

- 개념이 복잡하고 구현이 어려움

- 동적으로 새로운 노드를 삽입하거나 삭제하기가 간편하며, 연결 구조를 통해 물리 메모리를 연속적으로 사용하지 않아도 되기 때문에 메모리 사용 효율성이 높고 관리가 편함

- 연결 리스트는 배열과는 달리 특정 인덱스에 접근하기 위해서는 전체를 순서대로 읽어야 하므로 상수 시간에 접근할 수 없고 탐색에는 O(n) 소요, 아이템을 추가하거나 삭제, 추출하는 작업은 O(1) 가능, 즉 데이터 접근이 느림

- 연결리스트 종류 한방향연결리스트, 양방향 연결리스트,원형연결리스트

## 배열과 연결리스트의 차이

- 배열은 연속된 공간에 원소를 저장하는 반면 연결 리스트는 공간의 연속성이 없어, 정렬이나 검색 작업을 할 때 배열보다 연결리스트에서 시간이 더 걸린다, 배열은 i 번째 원소를 찾을 때 i 값으로 즉시 접근하고(O(1)) 연결리스트는 헤드노드에서부터 시작하여 i번 따라간 다음 i번 원소를 찾는다(O(i)).
- 연결 리스트에는 다음 원소의 링크를 위한 공간이 추가로 필요, 배열은 다음원소에 대하 링크가 없으므로 원소 하나당 필요한 공간이 상대적으로 더 작다.
- 배열은 들어올 원소의 총 수를 미리 알수 없는 경우가 대부분이라 충분한 수의 원소를 소화할 수 있는 크기로 선언해야 해서 공간 낭비가 불가피하거나 중간에 더 큰 배열을 새로 할당받아 원들을 옮겨야 하는 불편도 있다.
- 배열은 시작부터 고정된 크기를 지정하고, 연결리스트는 원소가 들어오는 대로 공간을 할당받아 넣는다, 즉 배열은 정적이고 연결 리스트는 동적이라고 할 수 있다.

## 연결 리스트의 객체 구조

- 원소를 저장하는 item 필드와 다음 노드를 가리키는 next 필드로 구성(데이터와 포인터)
- item 필드는 정수나 실수 같은 숫자타입이나 복잡한 객체가 될 수 있고, next 필드는 다음 노드를 가리키는 링크

image.png

- 연결리스트의 기본 형태는 첫번째 노드를 가리키는 레퍼런스 `__`head 있고 나머지 노드는 첫번째 노드에서 링크를 따라 접근하며 맨 마지막 노드의 링크는 뒤에 더 이상 노드가 없다는 의미로 None 값을 할당(/표시)

image.png

- 연결리스트에 대한 접근은 첫번째 노드부터 시작하는데, 리스트가 링크로 쭉 연결되어 있어 원소를 찾을 때, 새 원소를 삽입할 때, 원소를 삭제할 때 등 거의 모든 작업이 첫번째 노드를 거쳐간다. 첫번째 노드를 밟은 다음 링크를 타고 그 다음 노드들에 차례로 접근
- 연결리스트 객체는 리스트의 노드들을 직접 저장할 필요가 없고 리스트의 첫번째 노드에 접근할 수 있는 레퍼런스만 갖고 있으면 됨

# 탐욕법(Greedy)
    
문제를 해결하는 방법 중 하나로, 매 선택 시점마다 가장 최적이라고 생각되는 것을 선택해 나가는 방식입니다. 이 방법은 항상 전체적으로 가장 최적의 해를 보장하지는 않지만, 특정 종류의 문제에서는 효율적이고 정확한 해법이 될 수 있습니다

## 탐욕법 개념

- 최적의 해를 구하는데 사용되는 알고리즘 설계기법 중 하나로 **탐욕** 알고리즘 또는 **욕심쟁이** 알고리즘 또는 **그리디** 알고리즘은 이라고 한다.
- 이 알고리즘은 매 단계에서 현재 상태에서 가장 최적인 선택을 함으로써 전체 문제의 최적해에 도달할 수 있을 때 즉, 현재의 최선의 선택이 마지막 해답의 최적성을 해치지 않을때 사용하는 방법(즉 최적의해 보장하지 않음)
-그리디 알고리즘을 이용하면 매 순간 가장 좋아 보이는 것만 선택하며, 현재의 선택이 나중에 미칠 영향에 대해서는 고려하지 않습니다.
