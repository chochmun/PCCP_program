### 자료구조

#### 자료구조(Data Structure) 란?


- 데이터를 효율적으로 저장하고 관리하기 위해 사용하는 구조적인 방법
- 데이터를 처리하기 위해 모든 데이터는 메모리에 저장되어야 하는데 프로그래밍에서 데이터를 구조적으로 저장하는 방식 또는 이때 필요한 알고리즘
- 자료구조라하면 데이터와 그 데이터에 행할 수 있는 연산

#### 자료구조의 종류와 분류

##### 선형 vs. 비선형



image.png

- 선형자료구조
 - 배열, 연결리스트, 스택, 큐와 같이 순서가 정해져 있는 자료구조
 - 데이터를 정해진 자료형으로 순서대로 저장
 - 여러 개의 데이터 중에서 특정 데이터에 순차적으로 접근 할 수 있고, 접근하는 방법을 제시

- 비선형 자료구조
 - 트리와 그래프와 같이 순서가 정해지지 않은 자료구조
 - 데이터를 정해진 자료형으로 순서와 상관없이 저장
 - 기본 자료구조를 다양하게 적용하여 표현
 - 저장된 여러 개의 데이터 중에 특정 데이터에 접근하려면 저장된 규칙을 잘 이해하고 알아야 접근가능

##### 기본자료구조 vs. 확장자료구조

image.png

- 기본자료구조
 - 실제 주기억장치에 저장되는 방법을 말함

- 확장자료구조
 - 추상적 자료구조로 기본자료구조에 추상적으로 특정한 규칙을 포함시킨 자료구조
 - 스택, 큐, 트리, 그래프 등을 실제 구현할 때는 다음과같이 각각의 자료규칙에 맞게 기본 자료구조로 저장하는 것을 말함    
 예) 배열로 구현한 스택, 연결리스트로 구현한 큐
 - 기본 자료구조를 바탕으로 특정한 규칙(알고리즘)을 프로그래밍하여 구현

### 알고리즘

#### 알고리즘(Algorithm) 이란?

- [사전적정의] 알고리즘은 어떤 문제를 해결하기 위한 절차, 방법, 명령어들의 집합
- [프로그래밍] 주어진 문제의 해결을 위한 자료구조와 연산방법에 대한 선택
- 컴퓨터 알고리즘이란 문제를 컴퓨터를 통하여 해결하고자 할 때 컴퓨터가 처리할 일 또는 명령어의 순서


#### 알고리즘의 조건

- 입력: 0개 이상의 입력 존재
- 출력: 1개 이상의 출력 존재
- 명확성: 각 명령의 의미는 모호하지 않고 명확해야 함
- 유한성: 한정된 수의 단계 후에는 반드시 종료되어야 함
- 유효성: 각 명령어들은 실행가능한 연산 이어야 함


#### 알고리즘의 기술 방법

- 한국어와 영어와 같은 자연어
- 순서도
- 유사코드
- c와 파이썬 과 같은 프로그래밍 언어

#### 알고리즘의 성능

- 알고리즘의 효율성은 크게 두가지로 기준(시간적 효율성, 공간적 효율성)으로 측정하는데 일반적으로 시간적 효율성이 공간적 효율성보다 더욱 강조
  - 시간의 효율성: 문제 해결을 위해 얼마나 많은 시간을 필요로 하는가?
  - 공간적 효율성: 문제 해결을 위해 얼마나 많은 메모리 공간을 필요로 하는가
  


##### 알고리즘 수행 시간

- 알고리즘의 수행시간은 입력의 크기에 대해 시간이 얼마나 걸리는지로 표현
- **입력크기에 따라서 수행되는 연산횟수를 의미**


**[알고리즘 1]**
```
sample1(A[], n):
        k ← n/2
        return A[k]
```
  - 입력 배열의 크기 n에 상관없이 일정한 시간(상수시간)이 소요

**[알고리즘 2]**
```
sample2(A[], n):
        sum ← 0
        for i ← 0 to n-1
                sum ← sum + A[i]
        return sum
```
  - 배열의 모든 원소를 더하는 알고리즘으로 for 루프를 제외한 나머지 부분은 상수 시간이 소요되므로 for루프가 시간을 지배, for 루프 관련 수행 시간은 n에 비례

**[알고리즘 3]**
```
sample3(A[], n):
        sum ← 0
        for i ← 0 to n-1
                for j ← 0 to n-1
                        sum ← sum + A[i] * A[j]
        return sum
```
  - 배열의 모든 원소 쌍의 곱을 합산하는 알고리즘으로 for 루프가 중첩되어 총 n*n 번 반복되어  알고리즘의 총 수행시간은 $ n^2 비례$

**[알고리즘 4]**
```
sample4(A[], n):
        sum ← 0
        for i ← 0 to n-1
                for j ← 0 to n-1
                        k ← A[0...n-1]에서 임의로 n/2개를 뽑은 것들 중 최댓값
                        sum ← sum + k
        return sum
```
 - 크기가 n인 배열에서 임의로 n/2개를 뽑은 것들 중 최댓값 구하는 것은 n/2 시간에 비례, 즉 n에 비례
 - for 루프의 반복 횟수와 위의 수행시간 계산하면 $ n^2 * n = n^3 비례$

**[알고리즘 5]**
```
factorial(n):
        if (n = 1) return 1
        return n* factorial(n- 1)
```
 - 함수 factorial()이 몇번 호출되었는 지가 시간을 좌우, n번 호출되었으므로 n 비례




##### 알고리즘 복잡도(Complexity of Algorithms)

- 알고리즘이 실행함에 있어, 문제의 크기 (일반적으로 데이터 원소의 개수를 뜻합니다) 가 커짐에 따라서 얼마나 큰 시간을 (또는 공간을) 요구하느냐를 뜻
    - 시간의 복잡도(time complexity): 문제의 크기와 이를 해결하는 데 걸리는 시간 사이의 관계

    -공간 복잡도(Space Complexity) 문제의 크기와 이를 해결하는 데 필요한 메모리 공간 사이의 관계
- 알고리즘의 복잡도를 나타낼 때는 점근적 표기를 쓰는데, 점근적 복잡도란 입력의 크기가 충분히 클때의 복잡도, 즉 입력값 n이 무한대를 향할 때 함수의 실행 기간의 추이를 의미
  
   (입력의 크기가 작으면 복잡한 알고리즘이든 효율적인 알고리즘이든 실제 수행 시간은 별 차이가 없다)

###### 점근적 복잡도 표기법

 - O(빅오) 표기: 점근적 상한(시간의 복잡도 함수의 가장 높은 차수), 최악
 - 𝝮(오메가) 표기: 점근적 하한, 최선
 - 𝚹(세타) 표기: 점근적 동일, 평균

**빅오**

- 빅오는 점근적 실행 시간을 표기할 때 가장 널리 쓰이는 수학적 표기법 중 하나로 상한을 의미
- 빅오로 시간 복잡도를 표현할 때 최고차항만을 표기하며, 계수는 무시
- 알고리즘은  시간과 공간이 트레이드 오프 관계인 경우가 많음

    - 입력 패턴에 따라 정렬 속도에 차이가 있지만 정렬문제에 대해 O(nlogn) 보다 낮은 복잡도를 갖는 알고리즘은 존해할 수 없음이 증명되어 있음

###### 빅오 표기법의 다양한 예와 비교

**빅오 표기법의 예**

- O(1): 입력값이 아무리 커도 실행시간은 일정, 최고의 알고리즘이라 할 수 있음
- O(log n): 로그는 매우 큰 입력값에도 크게 영향을 받지 않는 편으로 웬만한 n의 크기에 대해서도 매우 견고
- O(n): 입력값만큼 실행 시간에 영향을 받으며, 알고리즘을 수행하는데 걸리는 시간은 입력값에 비례, 이러한 알고리즘을 선형시간 알고리즘이라고 함.정렬되지 않은 리스트에서 최대, 최소값을 찾는 경우가 이에 해당
- O(n log n): 병합정렬을 비롯한 대부분의 효율 좋은 정렬 알고리즘에 해당. 적어도 모든 수에 대해 한번 이상을 비교해야 하는 비교기반 정렬 알고리즘은 아무리 좋은 알고리즘도 O(n log n)보다 빠를 수 없음
- O($n^2$): 버블정렬같은 비효율적인 정렬알고리즘이 이에 해당
- O($2^n$): 피보나치 수를 재귀로 계산하는 알고리즘이 이에 해당
- O(n!): 가장 느린 알고리즘으로 입력값이 조금만 커져도 웬만한 다항시간 내에는 계산이 어려움 (각 도시를 방문하고 돌아오는 가장 짧은 경로를 찾는 외판원 문제를 브루트 포스로 풀이할 때 해당)

## 실습문제02 (크기가 작은 부분 문자열)

````python
def solution(t, p):
    t_lengh = len(t)
    p_length = len(p)
    cnt=0

    for i in range(t_lengh - p_length +1):
        if t[i:i+p_length] <= p:
            cnt+=1

    answer = cnt
    return answer
````

````python
t = "99"
p = "100"

result_str = t <= p  # "99" <= "100"
print(result_str)  # False (사전순으로 "99"가 "100"보다 뒤에 있음)
result_int = int(t) <= int(p)  # 99 <= 100
print(result_int)  # True (수치적으로 99가 100보다 작음)
````

## 선형 배열(Linear Array)

- 배열은 연속된 메모리 공간에 순차적으로 데이터들이 저장되는 자료구조

 (낭비되는 공간이  거의 없지만 큰배열일 경우 필요 메모리할당이 불가능할 수도 있음)
- 대부분의 프로그래밍 언어에서 동일 타입의 데이터를 저장
- 배열을 구성하는 각각의 값을 요소(element)라고 하고 배열에서 위치를 가리키는 숫자를 인덱스(index)라고 하고 인텍스를 통해 접근 가능. 즉 indexing과 slicing 가능

## 장점 & 단점

장점

- 배열의 각 요소에 접근하는 시간은 O(1)로 모두 동일, 검색 빠름

 (기본위치 + 오프셋(요소크기 * 인덱스) 연산으로 모든 요소에 접근 가능)
- 간단하고 사용하기 쉬움
- 기록밀도가 1이기 때문에 공간의 낭비가 적음(연결리스트는 포인터등의 부가정보를 가짐)

````python
# 리스트 초기화
lst = [1, 2, 3, 4, 5]

# append(): 리스트의 끝에 항목을 추가함.
lst.append(6)
print(lst)  # [1, 2, 3, 4, 5, 6]

# clear(): 리스트의 모든 항목을 제거함.
lst.clear()
print(lst)  # []

# copy(): 리스트의 얕은 복사본을 만듦.
lst = [1, 2, 3, 4, 5]
new_lst = lst.copy()
print(new_lst)  # [1, 2, 3, 4, 5]

# count(): 리스트에서 특정 항목이 나타나는 횟수를 반환함.
count = lst.count(2)
print(count)  # 1

# extend(): 리스트의 끝에 다른 리스트의 항목들을 추가함.
lst.extend([6, 7, 8])
print(lst)  # [1, 2, 3, 4, 5, 6, 7, 8]

# index(): 리스트에서 특정 항목을 찾아 첫 등장하는 인덱스를 반환함.
index = lst.index(4)
print(index)  # 3

# insert(): 지정된 위치에 항목을 삽입함.
lst.insert(3, 'a')
print(lst)  # [1, 2, 3, 'a', 4, 5, 6, 7, 8]

# pop(): 지정된 위치의 항목을 리스트에서 제거하고, 그 항목을 반환함.
item = lst.pop(3)
print(item)  # 'a'
print(lst)  # [1, 2, 3, 4, 5, 6, 7, 8]

# remove(): 리스트에서 첫 번째로 등장하는 특정 항목을 제거함.
lst.remove(8)
print(lst)  # [1, 2, 3, 4, 5, 6, 7]

# reverse(): 리스트의 항목들을 역순으로 뒤집음.
lst.reverse()
print(lst)  # [7, 6, 5, 4, 3, 2, 1]

# sort(): 리스트의 항목들을 정렬함.
lst.sort()
print(lst)  # [1, 2, 3, 4, 5, 6, 7]

# 역정렬
lst.sort(reverse=True)
print(lst)  # [7, 6, 5, 4, 3, 2, 1]
````

#### 연습문제 (2)

주어진 리스트에서 특정 원소를 (여러 번 들어 있을 수 있음) 모두 찾아내라.(인덱스 찾기)  

힌트:  
(1) index() 메서드를 이용한다.  
   찾는값 없으면 ValueErorr 발생
(2) 리스트 슬라이싱을 이용해서 반복하여 찾는다.  
  

````python
def find_all_occurrences(lst, value):
  init_index = 0
  result = []
  while True:
    try:
      index = lst.index(value, init_index)
      result.append(index)
      init_index = index + 1
    except ValueError:
      break
    pass
  return result



my_list = [1, 2, 3, 2, 4, 2, 5]
value_to_find = 2
result = find_all_occurrences(my_list, value_to_find)
print(result)  # 출력: [1, 3, 5]
````

## 탐색 (search) 이란?


- 복수의 원소로 이루어진 데이터에서 특정 원소를 찾아내는 작업
- 탐색 기본적인 방법

    - 선형 탐색 (linear search), 또는 순차 탐색 (sequential search): 순차적으로 모든 요소들을 탐색하여 원하는 값을 찾아냅니다. 배열의 길이에 비례하는 시간이 걸리므로, 최악의 경우에는 배열에 있는 모든 원소를 다 검사해야 할 수 있습니다.

    - 이진 탐색 (binary search): 탐색하려는 배열이 이미 정렬되어 있는 경우에만 적용할 수 있습니다. 배열의 가운데 원소와 찾으려 하는 값을 비교하면, (크기 순으로 정렬되어 있다는 성질을 이용하면) 왼쪽에 있을지 오른쪽에 있을지를 알 수 있습니다 (만약 있긴 있다면). 그러면, 적어도 반대쪽에 없는 것은 확실하므로, 배열의 반을 탐색하지 않고 버릴 수 있습니다. 이 과정을 반복하면 원하는 값을 찾아낼 수 있습니다.

이진 탐색이 선형 탐색보다 빠른 방법이기는 합니다. 그러나, 뭔가를 찾으려 한다고 할 때, 항상 이진 탐색 방법을 적용하는 것이 답인가요? 그러려면 우선 배열을 정렬해야 되어있어야 합니다.
